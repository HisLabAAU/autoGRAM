```python
import pandas as pd
import spacy

nlp = spacy.load("en_core_web_sm", disable=["parser", "lemmatizer", "ner"]) # First python -m spacy download en_core_web_sm

df = pd.read_csv("Run_ads_standardised.csv") # This is a slightly reworked version of https://worck.quarto.pub/datasets/Run_ads_croix.html with renamed columns, text, id, date
df = df.reset_index().rename(columns={"index": "doc_id"})
texts = df["text"].fillna("").astype(str)

rows = []
for doc_id, doc in zip(df["doc_id"], nlp.pipe(texts, batch_size=50)):
    for i, token in enumerate(doc):
        rows.append({
            "doc_id": doc_id,
            "token_index": i,
            "token": token.text,
            "pos": token.pos_,
            "tag": token.tag_,
            "start_char": token.idx,
            "end_char": token.idx + len(token.text),
        })

pos_long = pd.DataFrame(rows)
pos_long.to_csv("Run_ads_standardised_pos_long.csv", index=False)
```
